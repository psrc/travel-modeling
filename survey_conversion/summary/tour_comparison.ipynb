```python
# Compare new survey to old
import os
import toml
import pandas as pd
from functions import load_data, plot_display

config = toml.load(os.path.join(os.getcwd(), 'summary_configuration.toml'))

```

```python
print(config['survey_1_name']+': '+config['survey_1_dir'])
print(config['survey_2_name']+': '+config['survey_2_dir'])
```

```python
df = load_data('_tour.tsv', config['survey_1_dir'], config['survey_2_dir'], 
               config['survey_1_name'], config['survey_2_name'])
```

```python
for col in ['tour','pdpurp','tmodetp','tripsh1','tripsh2','tpathtp','toadtyp','tdadtyp']:
    plot_display(df, index=col, columns='source', aggfunc='sum', values='toexpfac')
```

```python

for col in ['tour','pdpurp','tmodetp','tripsh1','tripsh2','tpathtp','toadtyp','tdadtyp']:
    plot_display(df, index=col, columns='source', aggfunc='count', values='toexpfac')
```
