```python
import os
import pandas as pd
```

```python
output_dir = r'C:\Workspace\travel-modeling\aq_tool\output\mic\2018'

df = pd.DataFrame()

for fname in os.listdir(output_dir):
    _df = pd.read_csv(os.path.join(output_dir,fname))
    # Only include light, medium, and heavy emissions
    _df = _df[_df['veh_type'].isin(['light','medium','heavy'])]
    _df = _df[_df['pollutant_name'].isin(['CO', 'PM25 Total', 'NOx'])]
    _df['vmt'] = _df['interzonal_vmt'] + _df['intrazonal_vmt']
    _df = _df.groupby(['pollutant_name','veh_type']).sum()[['running_daily_tons','vmt']].reset_index()
    _df['running_daily_kg'] = _df['running_daily_tons']*1000
    _df['MIC'] = fname.split('.csv')[0]
    df = df.append(_df)
```

```python
df.to_clipboard()
```

```python
df_pivot = pd.pivot_table(df, index='MIC', columns=['pollutant_name','veh_type'], 
              values='running_daily_kg', aggfunc='sum')

df_pivot.to_csv('centers_emissions.csv')
df_pivot.to_clipboard()
```
