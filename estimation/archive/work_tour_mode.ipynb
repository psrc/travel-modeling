```python
import os
import pandas as pd
import toml

config = toml.load(os.path.join(os.getcwd(), 'coefficients_configuration.toml'))
```

```python
df_1 = pd.read_csv(os.path.join(config['dir_1'],'WorkTourMode\\WorkTourModeModel.F12'),
                    skiprows=3, delim_whitespace=True, nrows=35
                    )
df_1.columns = ['var_id','name','constrained','coef','pval']

df_2 = pd.read_csv(os.path.join(config['dir_2'],'WorkTourMode\\WorkTourModeModel.F12'),
                    skiprows=3, delim_whitespace=True, nrows=35
                    )
df_2.columns = ['var_id','name','constrained','coef','pval']
df = df_1.merge(df_2, suffixes=['_'+config['scen_1_name'], '_'+config['scen_2_name']],
                on=['var_id','name'])
df[['var_id','name','constrained_old','constrained_new','coef_old','coef_new',
    'pval_old','pval_new']]
```
