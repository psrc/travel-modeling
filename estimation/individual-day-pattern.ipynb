# Individual Day Pattern

```python
dir_14 = r'R:\SoundCastDocuments\2014Estimation\with_transit_sub_mode\results\IndividualDayPattern'
dir_06 = r'R:\SoundCastDocuments\2006Estimation\personbased_estimationfiles\newer_coeff_outputs'

fname_14 = 'IndividualPersonDayPatternModel.F12'
fname_06 = 'IndividualPersonDayPatternModel_psrcper1.F12'

df_14 = load_f12(dir_14, fname_14)
df_06 = load_f12(dir_06, fname_06)
```

```python
suffixes = ['_14','_06']
df = pd.merge(df_14, df_06, on='id', suffixes=suffixes)
df.index = df['coefficient_14']

# set resolution
figsize=(10,len(df)*.5)

field = 'value'
df[[field+suffix for suffix in suffixes]].plot(kind='barh',figsize=figsize, title='coefficient value', alpha=0.6)

field = 't-stat'
df[[field+suffix for suffix in suffixes]].astype('float').plot(kind='barh', figsize=figsize, title='t-stat',alpha=0.6)
```
