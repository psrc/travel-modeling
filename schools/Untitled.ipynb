```python
import pandas as pd
```

```python
enroll_df = pd.read_csv(r'R:\SoundCast\Inputs\2014_lodes\landuse\enrollment.csv')
```

```python
parcel = pd.read_csv(r'R:\SoundCast\Inputs\2014_lodes\landuse\parcels_urbansim.txt', delim_whitespace=True)
```

```python
# The fields on the original parcels file are incorrect
# Update with actual school enrollment values
school_cols = ['stugrd_p','stuhgh_p','stuuni_p']
parcel.drop(school_cols, axis=1, inplace=True)

# Merge
df = pd.merge(parcel,enroll_df,on='parcelid',how='left')
df[school_cols] = df[school_cols].fillna(0).astype('int')
```
