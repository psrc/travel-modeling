```python
import pandas as pd
```

# Heavy Trucks

```python
df = pd.read_csv(r"J:\Projects\INRIX\data\Truck volume by hour\heavy_trucks_volume_by_dow_hr.csv")
df = df[df['StartDOW'].isin(["Tuesday",'Wednesday','Thursday'])]
df = df.groupby(['StartTime_h']).sum()[['count']]/df['count'].sum()
```

```python
# Group by soundcast and activitysim TOD periods
df.loc[df.index.isin([20,21,22,23,24,0,1,2,3,4]), 'time_period'] = '20to5'
df.loc[df.index.isin([5]), 'time_period'] = '5to6'
df.loc[df.index.isin([6]), 'time_period'] = '6to7'
df.loc[df.index.isin([7]), 'time_period'] = '7to8'
df.loc[df.index.isin([8]), 'time_period'] = '8to9'
df.loc[df.index.isin([9]), 'time_period'] = '9to10'
df.loc[df.index.isin([10,11,12,13]), 'time_period'] = '10to14'
df.loc[df.index.isin([14]), 'time_period'] = '14to15'
df.loc[df.index.isin([15]), 'time_period'] = '15to16'
df.loc[df.index.isin([16]), 'time_period'] = '16to17'
df.loc[df.index.isin([17]), 'time_period'] = '17to18'
df.loc[df.index.isin([18,19]), 'time_period'] = '18to20'

```

```python
df_result = df.groupby('time_period').sum()[['count']]
df_result['model'] = 'daysim'
```

```python
# Process activitysim
df.loc[df.index.isin([20,21,22,23,24,0,1,2,3,4]), 'time_period'] = '20to5'
df.loc[df.index.isin([5,6,7,8]), 'time_period'] = '5to9'
df.loc[df.index.isin([9,10,11,12,13,14]), 'time_period'] = '9to15'
df.loc[df.index.isin([15,16,17]), 'time_period'] = '15to18'
df.loc[df.index.isin([18,19]), 'time_period'] = '18to20'
```

```python
df_asim = df.groupby('time_period').sum()[['count']]
df_asim['model'] = 'activitysim'

df_heavy_result = pd.concat([df_result, df_asim], axis=0)
df_heavy_result.reset_index(inplace=True)

df_heavy_result.rename(columns={'count': 'value'}, inplace=True)
```
