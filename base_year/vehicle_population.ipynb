```python
# Data Source: MOVES inputs from WA Department of Ecology, sourcetypeyear file by county
import pandas as pd
```

```python
year = '2023'
```

```python
input_dict = {
    'king': r'Y:\Air Quality\2023 Base Year\2023_MOVES_inputs\King\King_sourcetypeyear_2023.csv',
    'kitsap': r'Y:\Air Quality\2023 Base Year\2023_MOVES_inputs\Kitsap\Kitsap_sourcetypeyear_2023.csv',
    'pierce': r'Y:\Air Quality\2023 Base Year\2023_MOVES_inputs\Pierce\Pierce_sourcetypeyear_2023.csv',
    'snohomish': r'Y:\Air Quality\2023 Base Year\2023_MOVES_inputs\Snohomish\Snohomish_sourcetypeyear_2023.csv'
             }

results_df = pd.DataFrame()

for county, filedir in input_dict.items():
    df_county = pd.read_csv(filedir)
    county_df = pd.DataFrame()
    
    # https://github.com/USEPA/EPA_MOVES_Model/blob/master/docs/MOVES5CheatsheetOnroad.pdf
    # light vehicles
    df = df_county[df_county['sourceTypeID'].isin([11,21,31, 32])]    # Motorcycle, passenger car, passenger truck, light commercial truck
    df = pd.DataFrame(df[['sourceTypePopulation']].sum())
    df.rename(columns={0: 'vehicles'}, inplace=True)
    df['type'] = 'light'
    county_df = pd.concat([county_df, df])

    # medium truck
    df = df_county[df_county['sourceTypeID'].isin([51, 52, 53, 54])]    # refuse truck, single unit short- and long-haul truck, motor home
    df = pd.DataFrame(df[['sourceTypePopulation']].sum())
    df.rename(columns={0: 'vehicles'}, inplace=True)
    df['type'] = 'medium'
    county_df = pd.concat([county_df, df])

    # heavy truck
    df = df_county[df_county['sourceTypeID'].isin([61, 62])]    # combination short- and long-haul truck
    df = pd.DataFrame(df[['sourceTypePopulation']].sum())
    df.rename(columns={0: 'vehicles'}, inplace=True)
    df['type'] = 'heavy'
    county_df = pd.concat([county_df, df])

    county_df = county_df.reset_index(drop=True)
    county_df['county'] = county
    
    results_df = pd.concat([results_df, county_df])
results_df['year'] = year
```
