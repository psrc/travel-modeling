```python
import pandas as pd
```

```python
df = pd.read_csv(r'A:\aq\final_final.csv')
```

```python
# Join the exposure data to person information
```

```python
person = pd.read_csv(r'_person.tsv', sep='\t')
hh = pd.read_csv(r'_household.tsv', sep='\t')
```

```python
# Generate unique person ID field 
person['person_id'] = person['hhno'].astype('str') + '_' + person['pno'].astype('str')
```

```python
# Create ID Field
df = pd.merge(df,person,on='person_id',how='left')
```

```python
# Join hh info
df = pd.merge(df,hh,on='hhno',how='left')
```

```python
# df.groupby('')
# df[['hhincome','total_exposure']].groupby('hhincome').sum()
```

```python
# Create income classes 
df['inc_bin'] = pd.cut(df['hhincome'], bins=[0,10000,25000,50000,80000,150000,9999999], labels=['0-10,000','10,000-25,000','25,000-50,000',
                                                                             '50,000-80,000','80,000-150000','150,000+'])
```

```python
# Add geography to files
# RGCs?

# RGC based on home location
rgc = pd.read_csv(r'L:\T2040\soundcast_2014\scripts\summarize\inputs\parcels_in_urbcens.csv')
df = pd.merge(df,rgc,left_on='hhparcel',right_on='hhparcel',how='left')

# Rates by where people work

# Join RGC based on work location
df = pd.merge(df, rgc, left_on='pwpcl', right_on='hhparcel', how='left',suffixes=['_hh_rgc','_work_rgc'])
# person['pwpcl']
```

```python
df_positive = df[df['total_exposure'] > 0]
```

```python
by_res_type = pd.DataFrame(df_positive.groupby('hrestype').sum()['total_exposure']/df_positive.groupby('hrestype').count()['person_id'])
# hh.columns
```

```python
# 0?, 1: SFH, 2: Duplex, 3: Apartment, 4: Mobile home

# Average daily emissions total by residence type
by_res_type
```

```python
by_inc = pd.DataFrame(df_positive.groupby('inc_bin').sum()['total_exposure']/df_positive.groupby('inc_bin').count()['person_id'])
# hh.columns
```
