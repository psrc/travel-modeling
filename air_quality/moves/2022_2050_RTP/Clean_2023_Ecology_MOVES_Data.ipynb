```python
# Trim out unused sourceTypeIDs from county AVFT files
# These unused modes result in errors when imported to MOVES
import os
import pandas as pd

df_default = pd.read_csv(r'Y:\Air Quality\2026_2050_RTP\wa_dept_of_ecology_data\working_default_fuel_data\DEFAULT_avft.csv')
input_dir = r"Y:\Air Quality\2026_2050_RTP\wa_dept_of_ecology_data\2023_MOVES_inputs_original"
output_dir = r"Y:\Air Quality\2026_2050_RTP\wa_dept_of_ecology_data\2023_MOVES_inputs_revised"
```

```python
for county in ["King","Kitsap","Pierce","Snohomish"]:
    # Read in the county AVFT file
    df = pd.read_csv(os.path.join(input_dir, county, f"{county}_avft_2023.csv"))
    # Remove unused sourceTypeIDs as required for other fuel defaults
    df = df[df['sourceTypeID'].isin(df_default['sourceTypeID'].unique())]
    # Write out the revised AVFT file
    df.to_csv(os.path.join(output_dir, county, f"{county}_avft_2023.csv"), index=False)
```

```python
# Copy FuelUsageFraction into each county folder and update county ID
county_id_dict = {
    "King": 53033,
    "Kitsap": 53035,
    "Pierce": 53053,
    "Snohomish": 53061
}


for county, county_id in county_id_dict.items():
    # Read in the FuelUsageFraction file
    df = pd.read_csv(r"Y:\Air Quality\2026_2050_RTP\wa_dept_of_ecology_data\working_default_fuel_data\DEFAULT_FuelUsageFraction.csv")
    # Update the countyID column
    df['countyID'] = county_id
    # Write out the revised FuelUsageFraction file
    df.to_csv(os.path.join(output_dir, county, f"{county}_FuelUsageFraction_2023.csv"), index=False)
```
