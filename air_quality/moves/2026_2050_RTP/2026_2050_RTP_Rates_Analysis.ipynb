```python
import pandas as pd
```

```python
running_rates = pd.read_csv(r"Y:\Air Quality\2026_2050_RTP\moves_outputs\soundcast\running_emission_rates_by_veh_type.csv")
```

```python
# For a set speed, month, and pollutant, compare the year differences
df = running_rates.loc[
                  ((running_rates["pollutantID"] == 98) & 
                  (running_rates["monthID"] == 1) &
                  (running_rates['hourID'] == 7) &
                  (running_rates['veh_type'] == 'light') &
                  (running_rates['roadTypeID'] == 5)), 
                  ['year', 'ratePerDistance','avgSpeedBinID']
                  ]
# Plot ratePerDistance by speed bin for each year where year is shown as separate line
df = df.pivot(index='avgSpeedBinID', columns='year', values='ratePerDistance')
df.plot(title='Running Emission Rates for Light Vehicles at 7 AM in January (Pollutant ID 98, Road Type 5)', 
         xlabel='Average Speed Bin ID', ylabel='Rate Per Distance', marker='o')
```
