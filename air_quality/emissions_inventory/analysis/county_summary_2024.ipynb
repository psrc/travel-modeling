```python
import os
import pandas as pd
```

```python
def process_results(county, analysis_year, output_dir):  

    df_running = pd.read_csv(os.path.join(output_dir, f'interpolated/{county}/{analysis_year}/running_summary.csv'))
    df_start = pd.read_csv(os.path.join(output_dir, f'interpolated/{county}/{analysis_year}/start_summary.csv'))

    pollutant = 'CO2 Equivalent'
    df_pollutant_running = df_running[df_running['pollutant_name']==pollutant].groupby('mode').sum()[['running_daily_tons']]
    df_pollutant_start = df_start[df_start['pollutant_name']==pollutant].groupby('veh_type').sum()[['start_tons']]
    df_vmt = df_running[df_running['pollutant_name']==pollutant].groupby('mode').sum()[['daily_vmt_total']]

    return df_pollutant_running, df_pollutant_start, df_vmt
```

```python
county = 'King'
output_dir = 'Y:/Air Quality/King County Emissions Inventory/2024/data/output'
# analysis_year = '2024'
# veh_type_list = ['sov','hov2', 'hov3', 'medium', 'heavy', 'transit']

df_pollutant_running_23, df_pollutant_start_23, df_vmt_23 = process_results(county="King", analysis_year='2023', output_dir=output_dir)
```

```python
df_pollutant_running_24, df_pollutant_start_24, df_vmt_24 = process_results(county="King", analysis_year='2024', output_dir=output_dir)
```

```python
df_running = df_pollutant_running_23.merge(df_pollutant_running_24, left_index=True, right_index=True, suffixes=('_2023', '_2024'))
df_running = df_running.T[['sov','hov2','hov3','transit','medium','heavy']]
df_running.to_clipboard()
df_running
```

```python
df_start = df_pollutant_start_23.merge(df_pollutant_start_24, left_index=True, right_index=True, suffixes=('_2023', '_2024'))
df_start = df_start.T[["light","transit","medium","heavy"]]
df_start.to_clipboard()
df_start
```

```python
df_vmt = df_vmt_23.merge(df_vmt_24, left_index=True, right_index=True, suffixes=('_2023', '_2024'))
df_vmt = df_vmt.T[['sov','hov2','hov3','transit','medium','heavy']]
df_vmt.to_clipboard()
df_vmt
```
